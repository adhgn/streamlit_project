```python
from ucimlrepo import fetch_ucirepo

print("Downloading dataset directly...")
dataset = fetch_ucirepo(id=601)

# Combine features and targets into one dataframe
df = dataset.data.features.copy()
df['machine_failure'] = dataset.data.targets['Machine failure']

# Clean up column names to simple lowercase strings
rename_map = {
    'Air temperature [K]': 'air_temperature',
    'Process temperature [K]': 'process_temperature',
    'Rotational speed [rpm]': 'rotational_speed',
    'Torque [Nm]': 'torque',
    'Tool wear [min]': 'tool_wear'
}
df.rename(columns=rename_map, inplace=True)

# Save it right inside your backend directory
df.to_csv("backend/app/predictive_maintenance.csv", index=False)
print("Saved to backend/app/predictive_maintenance.csv successfully!")
```
